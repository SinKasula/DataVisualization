```python
from matplotlib import pyplot as plt
import pandas as pd
import plotly.graph_objects as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('./world-happiness-report-2019.csv')
df.head()
```

```python
df.columns
```

```python
df_select = df[['Country (region)', 'Social support', 'Freedom', 'Corruption','Generosity', 'Log of GDP\nper capita', 'Healthy life\nexpectancy']]
```

```python
def normalize_data(df, col):
    
    minval = df[col].min()
    maxval = df[col].max()
    meanval = df[col].mean()
    rangeval = maxval - minval 
    df[col] = (df[col] - meanval)/rangeval
    return df
    
```

```python
df_select = normalize_data(df_select,'Social support')
df_select = normalize_data(df_select,'Freedom')
df_select = normalize_data(df_select,'Generosity')
df_select = normalize_data(df_select,'Corruption')
df_select = normalize_data(df_select,'Log of GDP\nper capita')
df_select = normalize_data(df_select,'Healthy life\nexpectancy')
```

```python
df_select.head()
```

```python
df_select.describe()
```

```python
fig = go.Figure(data=
    go.Parcoords(
        line = dict(color = df_select.index,
                   colorscale = 'icefire',
                   showscale = True,
                   cmin = 0,
                   cmax = 250),
        dimensions = list([
            dict(range = [-0.5,0.5],
                 constraintrange = [-0.25,0],
                 label = 'Social Support', values = df_select['Social support']),
            dict(range = [-0.5,0.5],
                 label = 'Freedom', values = df_select['Freedom']),
            dict(range = [-0.5,0.5],
                 label = 'Corruption', values = df_select['Corruption']),
            dict(range = [-0.5,0.5],
                 visible = True,
                 label = 'Generosity', values = df_select['Generosity']),
            dict(range = [-0.5,0.5],
                 label = 'Log of GDP\nper capita', values = df_select['Log of GDP\nper capita']),
            dict(range = [-0.5,0.5],
                 label = 'Healthy life\nexpectancy', values = df_select['Healthy life\nexpectancy'])])
    )
)
fig.show()
```
